***Analysis***

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Age'],kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
```

**Distibution of respondents who have experienced domestic violence**

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df,x='Violence')
plt.title("Violence Distribution")
plt.xlabel("Experienced Domestic Violence")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(data=df,x='Violence',y='Age')
plt.title("Incidence of Domestic Violence by Age")
plt.xlabel("Experienced Domestic Violence")
plt.ylabel("Age")
plt.show()
```

```python
plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
sns.violinplot(data=df ,x='Violence',y='Age')
plt.title("Incidence of Domestic Violence by Age(Violin plot)")
plt.xlabel("Experienced Domestic violence")
plt.ylabel("Age")

plt.subplot(1,2,2)
sns.swarmplot(data=df , x='Violence', y='Age',palette='Set2')
plt.title("Incidence of Domestic Violence by Age( Swarm plot)")
plt.xlabel("Experienced Domestic Violence")
plt.ylabel("Age")
plt.show()

```

```python
# stored age
stored_age = df.sort_values(by="Age")

# occurence of age
age_count = stored_age["Age"].value_counts().sort_index()

age_countt = stored_age.groupby("Age")["Violence"].value_counts().unstack(fill_value=0)

# Display the stored age and corresponding Violence count
violence_count = pd.concat([age_count,age_countt],axis=1)
violence_count.columns=["Total","Violence","No Violence"]
violence_count
```

Relation Between Education Attianment and domestic violence

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x="Education",hue="Violence")
plt.title("Education Attainment and Domestic Violence")
plt.xlabel("Education Attainment")
plt.ylabel("Frequency")
plt.show()
```

Relationship Between Employment status And Domestic Violence

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df,x="Employment",hue="Violence")
plt.title("Employment Status And domestic Violence")
plt.xlabel("Employment Status")
plt.ylabel("Frequency")
plt.show()
```

Relationship Between Martial Status and Domestic Violence

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x="Marital status",hue="Violence")
plt.title("Martial status and Domestic Violence")
plt.xlabel("Martial Status")
plt.ylabel("Frequency")
plt.show()
```

Income and Domestic Violence

```python
plt.figure(figsize=(10,6))
sns.boxplot( data=df , x="Violence", y="Income")
plt.title("Income Distribution by Domestic Violence Experience")
plt.xlabel("Domestic Violence")
plt.ylabel("Income")
sns.despine(trim=True)
plt.show()
```

Income and domestic violence

```python
plt.figure(figsize=(10,6))
sns.stripplot(data=df,x="Violence",y="Income",hue="Marital status",jitter=True,dodge=True,size=5)
plt.title("Income Distribution by Marital Status and Domestic Violence Experience")
plt.xlabel("Domestic Violence")
plt.ylabel("Income")
plt.show()
```

```python
# count the Occurence of each unique salary
count_unique_salary  = df["Income"].value_counts()
# convert data series into dataframe
count_salary = pd.DataFrame(count_unique_salary.reset_index())
count_salary.columns=["Income","Counts"]
count_salary_sorted = count_salary.sort_values(by="Income")
count_salary_sorted
```

Income Distribution Analysis: Impact of Marital and Employment Status on Domestic Violence

```python

import matplotlib.gridspec as gridspec
fig = plt.figure(figsize=(15,10))
gs=gridspec.GridSpec(1,2,width_ratios=[1,1])

# income distribution by martial and domestic violence
ax1=plt.subplot(gs[0])
sns.stripplot(data=df,x="Marital status",y="Income",hue="Violence",jitter=True,dodge=True,size=5)
ax1.set_title("Income Distribution by Martial Status and Domestic Violence Experience")
ax1.set_xlabel("Marital Status")
ax1.set_ylabel("Income")
ax1.set_xticklabels(ax1.get_xticklabels(),rotation=45)
ax1.legend(title="Experienced Domestic Violence")
#  income Distribution by Employment and Violence
ax2=plt.subplot(gs[1])
sns.stripplot(data=df,x="Employment",y="Income",hue="Violence",jitter=True,dodge=True,size=5)
ax2.set_title("Income Distribution by Employment Status and Domestic Violence Experience")
ax2.set_xlabel("Employment Status")
ax2.set_ylabel("Income")
ax2.set_xticklabels(ax2.get_xticklabels(),rotation=45)
ax2.legend(title="Experienced Domestic Violence")

plt.show()
```
